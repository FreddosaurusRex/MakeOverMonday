# Slight Challenge
so in the next step i'm going to do a groupby on the dataframe, to create a lot of smaller dataframes associated by team. 

i thought i'd be able to plot them 'in line' so i could print out the plot in one short for loop. 

the work around would be filtering this by every team into it's own dataframe.. and then outputting it. as far as i can see. 

```python
%matplotlib inline
#groupby takes a column, and for every unique value in that column
#it creates another dataframe. in this case key is the team
#value is the resultant dataframe
groupedByTeam = inputDf.groupby("Team")
for key, value in groupedByTeam:
    print("DataFrame for the ",key)
    #this is annoying, Price has white spaces around the column name. 
    print(list(value.columns))
    pairplot = sb.jointplot("Year"," Price ",data=value, kind="reg")
    pairplot.plot
    
```
